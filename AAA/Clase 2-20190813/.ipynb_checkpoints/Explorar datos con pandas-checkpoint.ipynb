## Explorando los datos
Lo primero que vamos a ver es la funcion `describe`. Esta función de Pandas nos permite tener un acceso rápido a las principales características del dataset.

````python
df.describe()
````

Esa función crea una matriz de las variables junto con algunos datos estadísticos, como la media, la desviación estándar, mínimos y máximos, enetre otras; pero como vemos, no podemos ir más allá de un pequeño review de nuestras variables.

Otra forma de tener una visión general de los datos es utilizando la biblioteca [pandas-profiling](https://github.com/pandas-profiling/pandas-profiling). Dado un dataFrame, genera un reporte de los datos que contiene. Generar este reporte puede llevar bastante tiempo, de manera que podemos considerar tomar un subconjunto del dataset en vez de utilizarlo todo, con ayuda de la funcion [sample](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sample.html) de pandas:

```python
df_sampled = df.sample(frac=.5, replace=False, random_state=0)
```

El reporte contiene cinco secciones: _Overview, Variables, Correlations, Missing Values_ y _Sample_.
 * **Overview**: Esta sección nos muestra, datos generales del dataset como el tipo de datos y la cantidad de registros leidos, si hay filas duplicadas, o la memoria que ocupa el dataframe. Una parte muy útil es la sección de _Warnings_ en la que típicamente se informa de variables que tienen muchos ceros o nulos, y las variables que están altamente correlacionadas con otras, y que serán ignoradas (`rejected`) en el resto del reporte. 
 
 * **Variables**: La sección Variables muestra distintas estadísticas para cada una de las variables consideradas en el reporte, típicamente, mínimmos y máximos, valores frecuentes, cantidad de valores unicos. El botón _Toggle details_ muestra incluso más información, como cuantiles o histogramas.
 
 * **Correlations**: Esta sección muestra la matriz de correlación entre cada par de variables. Si la variable objetivo es parte del dataframe, también será incluido. Las más usuales son correlación de [Pearson](https://es.wikipedia.org/wiki/Coeficiente_de_correlaci%C3%B3n_de_Pearson) ($r$) y [Spearman](https://es.wikipedia.org/wiki/Coeficiente_de_correlaci%C3%B3n_de_Spearman) ($\rho$). 
 
 * **Missing values**: Esta sección es bastante sencilla, indica los valores faltantes en cada variable. Generalmente los valores faltantes son sustituidos por `np.nan`, y estos son los valores que se reportan en esta sección. Sin embargo, la forma en que se guardan los atributos vacios o desconocidos es arbitraria: a veces se usa el valor 0 o -1. en esos casos esta sección no los va a detectar, por lo que debemos prestar atención a las advertencias que brinda la sección **Overview** y a los valores mostrados en la sección **Variables**.
 
 * **Sample**: Esta última sección nos muestra las primeras y últimas filas del dataframe. Es exactamente lo mismo que hacer `df.head()` y `df.tail()`. Esto es útil sobre todo cuando trabajamos con datos temporales.

````python
report = df.profile_report(title='Reporte sobre datos disponibles', style={'full_width':True})
report
````

Además este reporte se puede exportar a un archivo html. Esto es muy útil para acceder al reporte sin necesidad de instalar nada.

````python
report.to_file(output_file="data/pandas_profile_report.html")
````

Por último, es interesante ver cómo se relacionan los atributos entre sí gráficamente. Notar que si el dataframe contiene a la variable a predecir, también vamos a ver cómo se relaciona esta con las demás variables. 

Cada entrada de la matriz $C_{i,j}$ muestra la relación que existe entre el atributo $i$ y el $j$, y para la diagonal $C_{i,i}$ muestra el histograma del atributo $i$. En este caso, el color indica el valor de la variable objetivo, donde el negro es el valor mínimo, y blanco el máximo.

````python
i = pd.plotting.scatter_matrix(df, c=df.target, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=.8)
````
